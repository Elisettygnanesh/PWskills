```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

def merge_sorted_lists(head1, head2):
    if head1 is None:
        return head2
    if head2 is None:
        return head1

    dummy_node = Node(0)
    merged_tail = dummy_node

    while head1 and head2:
        if head1.data <= head2.data:
            merged_tail.next = head1
            head1 = head1.next
        else:
            merged_tail.next = head2
            head2 = head2.next
        merged_tail = merged_tail.next


    if head1:
        merged_tail.next = head1
    elif head2:
        merged_tail.next = head2

    return dummy_node.next

def print_linked_list(head):
    current_node = head
    s=""
    while current_node:
        s+=str(current_node.data) + '-->' if current_node.next else str(current_node.data)
        current_node = current_node.next
    print(s)

list1_head = Node(1)
list1_head.next = Node(3)
list1_head.next.next = Node(5)
list1_head.next.next.next = Node(6)

list2_head = Node(2)
list2_head.next = Node(4)
list2_head.next.next = Node(7)
list2_head.next.next.next = Node(8)

print("First sorted linked list:")
print_linked_list(list1_head)

print("Second sorted linked list:")
print_linked_list(list2_head)

merged_head = merge_sorted_lists(list1_head, list2_head)

print("Merged sorted linked list:")
print_linked_list(merged_head)
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


def remove_nth_from_end(head, n):
    dummy = Node(0)
    dummy.next = head
    slow = fast = dummy

    for _ in range(n + 1):
        fast = fast.next

    while fast:
        slow = slow.next
        fast = fast.next

    slow.next = slow.next.next

    return dummy.next


def printl(head2):
    he = head2
    s=""
    while he:
        s+=str(he.data) + "-->" if he.next else str(he.data)
        he = he.next
    print(s)


head = Node(1)
head.next = Node(2)
head.next.next = Node(3)
head.next.next.next = Node(4)
head.next.next.next.next = Node(5)
head.next.next.next.next.next = Node(6)

n = 2
print("Original linked list:")
printl(head)

head = remove_nth_from_end(head, n)

print(f"After removing {n}th element from the end:")
printl(head)
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None


def find_inter(head1, head2):
    result = Node(0)
    r = result
    a = head1
    h = head2
    while a:
        b = head2
        while b:
            if a.data == b.data:
                ans = a.data
                r.next = Node(ans)
                r = r.next
                b = b.next
            else:
                b = b.next
        b = h
        a = a.next
    return result.next
    
def printl(head):
    m = head
    s=""
    while m:
        s+= str(m.data) + "-->" if m.next else str(m.data)
        m = m.next
    print(s)
    
values1 = [1, 2, 3, 4, 8, 6, 9]
head1 = Node(values1[0])
current = head1
for val in values1[1:]:
    current.next = Node(val)
    current = current.next
    
values2 = [5, 1, 6, 7]
head2 = Node(values2[0])
current = head2
for val in values2[1:]:
    current.next = Node(val)
    current = current.next

head0 = find_inter(head1, head2)
printl(head0)
```

```python
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

def addTwoNumbers(l1, l2):
    dummy = ListNode() 
    current = dummy
    carry = 0
    
    while l1 or l2 or carry:
        val1 = l1.val if l1 else 0
        val2 = l2.val if l2 else 0
        total = val1 + val2 + carry
        
        carry = total // 10
        
        current.next = ListNode(total % 10)
        current = current.next
        
        if l1:
            l1 = l1.next
        if l2:
            l2 = l2.next
    
    return dummy.next 

l1 = ListNode(2, ListNode(4, ListNode(3)))
l2 = ListNode(5, ListNode(6, ListNode(4)))

result = addTwoNumbers(l1, l2)
while result:
    print(result.val, end=" -> ")
    result = result.next
```

```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None
        self.arbit = None

def clone_linked_list(head):
    if not head:
        return None
    
    curr = head
    while curr:
        new_node = Node(curr.data)
        new_node.next = curr.next
        curr.next = new_node
        curr = new_node.next
    
    curr = head
    while curr:
        if curr.arbit:
            curr.next.arbit = curr.arbit.next
        curr = curr.next.next
    
    original_head = head
    clone_head = head.next
    clone_curr = clone_head
    
    while original_head:
        original_head.next = clone_curr.next
        original_head = original_head.next
        
        if original_head:
            clone_curr.next = original_head.next
            clone_curr = clone_curr.next
    
    return clone_head

original_head = Node(1)
original_head.next = Node(2)
original_head.next.next = Node(3)

original_head.arbit = original_head.next.next
original_head.next.arbit = original_head
original_head.next.next.arbit = original_head.next

cloned_head = clone_linked_list(original_head)

while cloned_head:
    print("Data:", cloned_head.data, "Arbit Data:", cloned_head.arbit.data if cloned_head.arbit else None)
    cloned_head = cloned_head.next
```
